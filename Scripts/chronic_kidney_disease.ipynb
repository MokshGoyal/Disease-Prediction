Importing the necessary libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
# Reading the dataset
df = pd.read_csv("Datasets/CKD_Preprocessed.csv")
df.head(5)
```

```python
# Checking if data has null values
df.isnull().sum()
```

```python
# Checking other descriptions of data
df.info()
```

```python
df.describe()
```

Since our dataset has no value so directly preprocessing data for K-nearest neighbours(KNN).

```python
X = df.drop('Chronic_Kidney_Disease',axis=1) 
y = df['Chronic_Kidney_Disease']
# X contains all the columns except the column Chronic Kidney disease
# y contains the column "Chronic Kidney disease"
```

```python
# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Splitting the data into train and test data
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y,test_size=0.2)
```

```python
# Creating object of KNN
knn_classifier = KNeighborsClassifier()
```

```python
# Training the model
knn_classifier.fit(X_train,y_train)
```

```python
# Evaluating the model(Checking accuracy)
accuracy = knn_classifier.score(X_test,y_test)
print("Accuracy of model: ",accuracy*100,"%")
```

```python
directory_path = 'saved_models/'
# Specify the filename
filename = 'CKD_prediction.sav'

# Concatenate directory path and filename to get the full path
file_path = directory_path + filename

# Open the file in binary write mode and dump the model
with open(file_path, 'wb') as f:
    pickle.dump(KNeighborsClassifier, f)
```
